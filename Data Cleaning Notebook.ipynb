```python
import pandas as pd

df = pd.read_excel("IMDB_data_dirty.xlsx")
df.head()
```

```python
# Cek jumlah duplikasi pada data
print("Jumlah duplikasi data: " + str(df.duplicated().sum()))
```

```python
# Cek missing values
missing_values = df.isnull().sum() / len(df) * 100
print("Missing Values:")
print(missing_values)
```

```python
# Drop kolom dengan missing values > 50%
drop_col = missing_values[missing_values > 50].index
df = df.drop(columns=drop_col)
print("Kolom yang di drop:", list(drop_col))
```

```python
skew = df.skew(numeric_only=True)
print(skew)
```

```python
# Handle missing values pada kolom numerikal dan kategorikal
df.fillna(df.median(numeric_only=True), inplace=True)
df.fillna(df.mode().iloc[0], inplace=True)
```

```python
# Cek missing values setelah dibersihkan
missing_values = df.isnull().sum() / len(df) * 100
print("Missing Values:")
print(missing_values)
```

```python
# Simpan hasil cleaning data
clean_df = "IMDB_data_clean.csv"
df.to_csv(clean_df, index=False)
```
